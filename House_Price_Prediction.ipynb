```python
# SalePrice
data_df['SalePrice'].hist(bins=50)
plt.title("Sale Pirce Distribution")
plt.show()
```

```python
# HISTOGRAM (Distribution of saleprice)
plt.figure(figsize=(10,5))
sns.histplot(data_df['SalePrice'],bins=50,kde=True)
plt.title('Distribution of SalePrice')
plt.show()
```

```python
# SCATTERPLOT(MS Zoning vs sale price)
plt.figure(figsize=(8,6))
sns.scatterplot(x='MS Zoning', y='SalePrice', data=data_df)
plt.title('MS Zoning vs SalePrice')
plt.show()
```

```python
# BAR PLOT (Street vs SalePrice)
plt.figure(figsize=(8,6))
sns.barplot(x='Street', y='SalePrice', data=data_df)
plt.title('Street vs SalePrice')
plt.show()
```

```python
# BOX PLOT(Alley vs SalePrice)
plt.figure(figsize=(8,6))
sns.boxplot(x='Alley', y='SalePrice', data=data_df)
plt.title('Alley vs SalePrice')
plt.show()
```

```python
# LINEAR REGRESSION 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

x = data_df[['Lot Area']]
y = data_df[['SalePrice']]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

plt.figure(figsize = (8,6))
plt.scatter(x_test, y_test, color = "blue", label = 'Actual SalePrice')
plt.plot(x_test, y_test, color = "red", linewidth = 2, label = 'Predicted')
plt.xlabel('Lot Area')
plt.ylabel('SalePrice')
plt.title('Linear Regression: Lot Area vs SalePrice')
plt.legend()
plt.show()
```
